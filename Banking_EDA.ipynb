```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Banking.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

# Step 1: Flatten any list values in 'Estimated Income' (e.g., [50000] âž 50000)
df['Estimated Income'] = df['Estimated Income'].apply(lambda x: x[0] if isinstance(x, list) else x)

# Step 2: Apply binning using pd.cut
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)

# Optional: View result
print(df[['Estimated Income', 'Income Band']].head())
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

# **Categorical Columns**

```python
# Examine the distribution of unique categories in categorical columns

categorical_cols = df[["BRId", "GenderId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification","IAId", "Properties Owned", "Risk Weighting","Income Band" ]]

for col in categorical_cols :
  print(f"Value Counts for '{col}': ")
  display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification","IAId", "Properties Owned", "Risk Weighting","Income Band" ]]):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)#, hue='GenderId')
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification","IAId", "Properties Owned", "Risk Weighting","Income Band" ]]):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue='Nationality')
```

# **Numerical Columns**

```python
df.head(1)
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account',  'Business Lending']

plt.figure(figsize=(15,15)) # (width, Height)
for i, col in enumerate(numerical_cols):
  plt.subplot(3,3, i+1) # (rows, columns)
  sns.histplot(df[col], kde=True)
  plt.title(col)
```

# **Heat Maps**

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
